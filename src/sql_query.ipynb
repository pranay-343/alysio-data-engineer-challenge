```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('../salesforce_data.db')
cursor = conn.cursor()

# Execute the SQL query
cursor.execute('''
SELECT 
    c.first_name,
    c.last_name,
    c.email,
    c.title,
    comp.name AS company_name,
    comp.industry,
    comp.annual_revenue
FROM Contacts c
INNER JOIN Companies comp ON c.company_id = comp.id
ORDER BY comp.annual_revenue DESC
''')

# Fetch all results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('../salesforce_data.db')
cursor = conn.cursor()

# Execute the SQL query
cursor.execute('''
SELECT 
    a.type as activity_type,
    a.outcome,
    COUNT(*) as activity_count,
    AVG(a.duration_minutes) as avg_duration,
    COUNT(DISTINCT a.contact_id) as unique_contacts,
    COUNT(DISTINCT o.id) as related_opportunities,
    SUM(o.amount) as total_opportunity_value
FROM Activities a
LEFT JOIN Opportunities o ON a.opportunity_id = o.id
GROUP BY a.type, a.outcome
ORDER BY activity_count DESC
''')

# Fetch all results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```
